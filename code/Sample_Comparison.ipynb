```python


# =========================================================
# 4) Effect size helper functions
# =========================================================

def hedges_g(mean_digi, sd_digi, n_digi, mean_psy, sd_psy, n_psy):
    """
    Positive g means PSYMPACT > DigiHero.
    """
    pooled_sd = sqrt(
        (((n_digi - 1) * sd_digi**2) + ((n_psy - 1) * sd_psy**2)) / (n_digi + n_psy - 2)
    )
    d = (mean_psy - mean_digi) / pooled_sd
    J = 1 - (3 / (4 * (n_digi + n_psy) - 9))  # small sample correction
    g = J * d
    return g

def cohens_h(p_digi, p_psy):
    """
    Positive h means higher proportion in PSYMPACT.
    """
    # protect against tiny numerical issues
    p_digi = min(max(p_digi, 0), 1)
    p_psy  = min(max(p_psy, 0), 1)
    return 2 * asin(np.sqrt(p_psy)) - 2 * asin(np.sqrt(p_digi))

def odds_ratio(yes_digi, no_digi, yes_psy, no_psy):
    """
    OR > 1 means higher odds in PSYMPACT.
    """
    return (yes_psy / no_psy) / (yes_digi / no_digi)

def cramers_v_from_table(table):
    """
    table: 2D numpy array of counts
    """
    table = np.asarray(table, dtype=float)
    n = table.sum()
    row_sums = table.sum(axis=1, keepdims=True)
    col_sums = table.sum(axis=0, keepdims=True)
    expected = row_sums @ col_sums / n
    chi2 = ((table - expected) ** 2 / expected).sum()
    r, c = table.shape
    return np.sqrt(chi2 / (n * min(r - 1, c - 1)))

def interpret_abs_effect(x, kind="generic"):
    x = abs(x)

    if kind == "g":
        if x < 0.20:
            return "negligible"
        elif x < 0.50:
            return "small"
        elif x < 0.80:
            return "medium"
        else:
            return "large"

    if kind == "h":
        if x < 0.20:
            return "negligible"
        elif x < 0.50:
            return "small"
        elif x < 0.80:
            return "medium"
        else:
            return "large"

    if kind == "v":
        if x < 0.10:
            return "negligible"
        elif x < 0.30:
            return "small"
        elif x < 0.50:
            return "medium"
        else:
            return "large"

    return np.nan

# =========================================================
# 5) Calculate effect sizes
# =========================================================

effect_rows = []

# -------------------------
# Continuous: Hedges' g
# -------------------------
for var in df_cont["variable"].unique():
    sub = df_cont[df_cont["variable"] == var].set_index("sample")

    m_digi = sub.loc["DigiHero", "mean"]
    sd_digi = sub.loc["DigiHero", "sd"]
    n_digi = sub.loc["DigiHero", "valid_n"]

    m_psy = sub.loc["PSYMPACT", "mean"]
    sd_psy = sub.loc["PSYMPACT", "sd"]
    n_psy = sub.loc["PSYMPACT", "valid_n"]

    g = hedges_g(m_digi, sd_digi, n_digi, m_psy, sd_psy, n_psy)

    effect_rows.append({
        "variable": var,
        "variable_type": "continuous",
        "effect_metric": "Hedges_g",
        "effect_size": g,
        "interpretation": interpret_abs_effect(g, kind="g"),
        "direction": "PSYMPACT > DigiHero" if g > 0 else "DigiHero > PSYMPACT",
        "digihero_valid_n": n_digi,
        "psympact_valid_n": n_psy
    })

# -------------------------
# Binary: OR + Cohen's h + phi/Cramér's V
# -------------------------
for var in df_bin["variable"].unique():
    sub = df_bin[df_bin["variable"] == var].set_index("sample")

    yes_digi = sub.loc["DigiHero", "yes_count"]
    no_digi  = sub.loc["DigiHero", "no_count"]
    p_digi   = sub.loc["DigiHero", "prop_yes"]

    yes_psy = sub.loc["PSYMPACT", "yes_count"]
    no_psy  = sub.loc["PSYMPACT", "no_count"]
    p_psy   = sub.loc["PSYMPACT", "prop_yes"]

    h = cohens_h(p_digi, p_psy)
    or_val = odds_ratio(yes_digi, no_digi, yes_psy, no_psy)
    v = cramers_v_from_table(np.array([
        [yes_digi, no_digi],
        [yes_psy, no_psy]
    ]))

    effect_rows.append({
        "variable": var,
        "variable_type": "binary",
        "effect_metric": "Cohen_h",
        "effect_size": h,
        "interpretation": interpret_abs_effect(h, kind="h"),
        "direction": "PSYMPACT > DigiHero" if h > 0 else "DigiHero > PSYMPACT",
        "digihero_valid_n": sub.loc["DigiHero", "valid_n"],
        "psympact_valid_n": sub.loc["PSYMPACT", "valid_n"]
    })

    effect_rows.append({
        "variable": var,
        "variable_type": "binary",
        "effect_metric": "Odds_ratio",
        "effect_size": or_val,
        "interpretation": "higher odds in PSYMPACT" if or_val > 1 else "higher odds in DigiHero",
        "direction": "PSYMPACT > DigiHero" if or_val > 1 else "DigiHero > PSYMPACT",
        "digihero_valid_n": sub.loc["DigiHero", "valid_n"],
        "psympact_valid_n": sub.loc["PSYMPACT", "valid_n"]
    })

    effect_rows.append({
        "variable": var,
        "variable_type": "binary",
        "effect_metric": "Phi_or_CramersV",
        "effect_size": v,
        "interpretation": interpret_abs_effect(v, kind="v"),
        "direction": np.nan,
        "digihero_valid_n": sub.loc["DigiHero", "valid_n"],
        "psympact_valid_n": sub.loc["PSYMPACT", "valid_n"]
    })

# -------------------------
# Multi-category: Cramér's V
# -------------------------
for var in df_cat["variable"].unique():
    pivot = (
        df_cat[df_cat["variable"] == var]
        .pivot(index="sample", columns="category", values="count")
        .loc[["DigiHero", "PSYMPACT"]]
    )

    v = cramers_v_from_table(pivot.to_numpy())

    effect_rows.append({
        "variable": var,
        "variable_type": "categorical",
        "effect_metric": "Cramers_V",
        "effect_size": v,
        "interpretation": interpret_abs_effect(v, kind="v"),
        "direction": np.nan,
        "digihero_valid_n": int(df_cat[df_cat["variable"] == var].query("sample == 'DigiHero'")["valid_n"].iloc[0]),
        "psympact_valid_n": int(df_cat[df_cat["variable"] == var].query("sample == 'PSYMPACT'")["valid_n"].iloc[0])
    })

df_effects = pd.DataFrame(effect_rows)

# =========================================================
# 6) Optional: sort for readability
# =========================================================
metric_order = {
    "Hedges_g": 1,
    "Cohen_h": 2,
    "Odds_ratio": 3,
    "Phi_or_CramersV": 4,
    "Cramers_V": 5
}
df_effects["metric_order"] = df_effects["effect_metric"].map(metric_order)
df_effects = df_effects.sort_values(["variable", "metric_order"]).drop(columns="metric_order")

# =========================================================
# 7) Quick checks
# =========================================================

# Check that category counts sum to valid_n for categorical variables
cat_check = (
    df_cat.groupby(["variable", "sample"])
    .agg(count_sum=("count", "sum"), valid_n=("valid_n", "first"))
    .reset_index()
)
cat_check["ok"] = cat_check["count_sum"] == cat_check["valid_n"]

# Check that yes + no = valid_n for binary variables
bin_check = df_bin.copy()
bin_check["ok"] = (bin_check["yes_count"] + bin_check["no_count"]) == bin_check["valid_n"]

print("\n=== Binary checks ===")
print(bin_check[["variable", "sample", "yes_count", "no_count", "valid_n", "ok"]])

print("\n=== Categorical checks ===")
print(cat_check)

print("\n=== Summary dataset ===")
print(df_summary.head(20))

print("\n=== Effect sizes ===")
print(df_effects)

# =========================================================
# 8) Save outputs
# =========================================================
df_summary.to_csv("summary_dataset_from_table.csv", index=False)
df_effects.to_csv("effect_sizes_from_table.csv", index=False)

print("\nSaved:")
print("- summary_dataset_from_table.csv")
print("- effect_sizes_from_table.csv")
```

## Compare Side Effects of Psympact and External

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind_from_stats

# =========================================================
# 1) Summary statistics
# =========================================================

# Psympact
psympact = {
    "dependency": {
        "mean": 0.451613,
        "sd": 0.731464,
        "n": 1178,
        "min": 0,
        "max": 3,
    },
    "decision_difficulty": {
        "mean": 0.280985,
        "sd": 0.603019,
        "n": 1178,
        "min": 0,
        "max": 3,
    },
    "social_impact": {
        "mean": 0.922619,
        "sd": 0.347585,
        "n": 1176,
        "min": 0,
        "max": 3,
    },
}

# External sample
external = {
    "dependency": {           # item 11
        "mean": 0.25,
        "sd": 0.54,
        "n": 1773,
        "min": 0,
        "max": 3,
    },
    "decision_difficulty": {  # item 12
        "mean": 0.26,
        "sd": 0.61,
        "n": 1773,
        "min": 0,
        "max": 3,
    },
    "social_impact": {        # item 7
        "mean": 0.11,
        "sd": 0.43,
        "n": 1773,
        "min": 0,
        "max": 3,
    },
}

# =========================================================
# 2) Helper functions
# =========================================================

def cohen_d_from_summary(m1, sd1, n1, m2, sd2, n2):
    pooled_sd = np.sqrt(
        ((n1 - 1) * sd1**2 + (n2 - 1) * sd2**2) / (n1 + n2 - 2)
    )
    return (m1 - m2) / pooled_sd

def hedges_g_from_summary(m1, sd1, n1, m2, sd2, n2):
    d = cohen_d_from_summary(m1, sd1, n1, m2, sd2, n2)
    J = 1 - (3 / (4 * (n1 + n2) - 9))
    return d * J

def effect_size_label(g):
    ag = abs(g)
    if ag < 0.20:
        return "very small"
    elif ag < 0.50:
        return "small"
    elif ag < 0.80:
        return "medium"
    else:
        return "large"

# =========================================================
# 3) Run comparisons
# =========================================================

rows = []

for var in ["dependency", "decision_difficulty", "social_impact"]:
    ps = psympact[var]
    ex = external[var]

    # Welch t-test from summary stats
    t_res = ttest_ind_from_stats(
        mean1=ps["mean"], std1=ps["sd"], nobs1=ps["n"],
        mean2=ex["mean"], std2=ex["sd"], nobs2=ex["n"],
        equal_var=False
    )

    # Hedges' g
    g = hedges_g_from_summary(
        ps["mean"], ps["sd"], ps["n"],
        ex["mean"], ex["sd"], ex["n"]
    )

    rows.append({
        "variable": var,
        "psympact_n": ps["n"],
        "external_n": ex["n"],
        "psympact_mean": ps["mean"],
        "psympact_sd": ps["sd"],
        "external_mean": ex["mean"],
        "external_sd": ex["sd"],
        "mean_diff_ps_minus_ext": ps["mean"] - ex["mean"],
        "welch_t": t_res.statistic,
        "p_value": t_res.pvalue,
        "hedges_g": g,
        "effect_size": effect_size_label(g),
    })

results = pd.DataFrame(rows)

# =========================================================
# 4) Pretty formatting
# =========================================================

pd.set_option("display.float_format", lambda x: f"{x:.4f}")
print(results)

print("\n" + "=" * 80)
print("APA-style summary")
print("=" * 80)

for _, row in results.iterrows():
    print(
        f"{row['variable']}: "
        f"Psympact (M = {row['psympact_mean']:.3f}, SD = {row['psympact_sd']:.3f}, n = {int(row['psympact_n'])}) vs. "
        f"External (M = {row['external_mean']:.3f}, SD = {row['external_sd']:.3f}, n = {int(row['external_n'])}); "
        f"Welch's t = {row['welch_t']:.3f}, p = {row['p_value']:.6g}, "
        f"Hedges' g = {row['hedges_g']:.3f} ({row['effect_size']})."
    )
```

```python
import numpy as np
import pandas as pd
from math import log, exp, sqrt
from scipy.stats import fisher_exact

# =========================================================
# 1) Input data
# =========================================================

PSYMPACT_N = 1180

psympact = {
    "dependency": {
        "n_elevated": 386,
        "n_total": PSYMPACT_N,
    },
    "decision_difficulty": {   # reduced decision-making autonomy
        "n_elevated": 247,
        "n_total": PSYMPACT_N,
    },
    "social_impact": {
        "n_elevated": 27,
        "n_total": PSYMPACT_N,
    },
}

# External sample:
# item 11 = dependency
# item 12 = decision_difficulty
# item 7  = social_impact
#
# n_total is back-calculated from the published count + rounded percentage.
# Replace with exact denominators if you have them.
external = {
    "dependency": {
        "itemno": 11,
        "n_elevated": 223,
        "n_total": 1517,   # approx from 223 / 14.7%
    },
    "decision_difficulty": {
        "itemno": 12,
        "n_elevated": 73,
        "n_total": 1521,   # approx from 73 / 4.8%
    },
    "social_impact": {
        "itemno": 7,
        "n_elevated": 20,
        "n_total": 1538,   # approx from 20 / 1.3%
    },
}

# =========================================================
# 2) Helper functions
# =========================================================

def odds_ratio_ci(a, b, c, d):
    """
    2x2 table:
                  Elevated   Not elevated
    PSYMPACT         a            b
    External         c            d
    """
    # continuity correction only if needed
    a_cc, b_cc, c_cc, d_cc = a, b, c, d
    if min(a, b, c, d) == 0:
        a_cc += 0.5
        b_cc += 0.5
        c_cc += 0.5
        d_cc += 0.5

    or_val = (a_cc * d_cc) / (b_cc * c_cc)
    se = sqrt(1/a_cc + 1/b_cc + 1/c_cc + 1/d_cc)

    lcl = exp(log(or_val) - 1.96 * se)
    ucl = exp(log(or_val) + 1.96 * se)
    return or_val, lcl, ucl

def risk_difference_ci(a, n1, c, n2):
    p1 = a / n1
    p2 = c / n2
    rd = p1 - p2
    se = sqrt((p1 * (1 - p1) / n1) + (p2 * (1 - p2) / n2))
    lcl = rd - 1.96 * se
    ucl = rd + 1.96 * se
    return rd, lcl, ucl

def nnh_from_rd(rd, rd_lcl, rd_ucl):
    """
    For harmful exposure when RD > 0:
    NNH = 1 / RD
    CI is inverted from RD CI.
    """
    if rd <= 0 or rd_lcl <= 0 or rd_ucl <= 0:
        return np.nan, np.nan, np.nan

    nnh = 1 / rd
    nnh_lcl = 1 / rd_ucl   # smaller NNH = greater harm
    nnh_ucl = 1 / rd_lcl
    return nnh, nnh_lcl, nnh_ucl

# =========================================================
# 3) Run comparisons
# =========================================================

rows = []

for var in ["dependency", "decision_difficulty", "social_impact"]:
    ps = psympact[var]
    ex = external[var]

    a = ps["n_elevated"]
    b = ps["n_total"] - ps["n_elevated"]
    c = ex["n_elevated"]
    d = ex["n_total"] - ex["n_elevated"]

    p1 = a / ps["n_total"]
    p2 = c / ex["n_total"]

    or_val, or_lcl, or_ucl = odds_ratio_ci(a, b, c, d)
    _, fisher_p = fisher_exact([[a, b], [c, d]], alternative="two-sided")

    rd, rd_lcl, rd_ucl = risk_difference_ci(a, ps["n_total"], c, ex["n_total"])
    nnh, nnh_lcl, nnh_ucl = nnh_from_rd(rd, rd_lcl, rd_ucl)

    rows.append({
        "variable": var,
        "psympact_n": ps["n_total"],
        "psympact_elevated": a,
        "psympact_pct": p1 * 100,
        "external_n": ex["n_total"],
        "external_elevated": c,
        "external_pct": p2 * 100,
        "or": or_val,
        "or_lcl": or_lcl,
        "or_ucl": or_ucl,
        "risk_diff": rd * 100,
        "risk_diff_lcl": rd_lcl * 100,
        "risk_diff_ucl": rd_ucl * 100,
        "nnh": nnh,
        "nnh_lcl": nnh_lcl,
        "nnh_ucl": nnh_ucl,
        "fisher_p": fisher_p,
    })

results = pd.DataFrame(rows)

# =========================================================
# 4) Pretty output
# =========================================================

pd.set_option("display.float_format", lambda x: f"{x:.4f}")
print(results)

print("\n" + "=" * 100)
print("Formatted results")
print("=" * 100)

label_map = {
    "dependency": "Dependency",
    "decision_difficulty": "Reduced decision-making autonomy",
    "social_impact": "Negative social impact",
}

for _, row in results.iterrows():
    label = label_map[row["variable"]]
    txt = (
        f"{label}: "
        f"{int(row['psympact_elevated'])}/{int(row['psympact_n'])} "
        f"({row['psympact_pct']:.1f}%) vs. "
        f"{int(row['external_elevated'])}/{int(row['external_n'])} "
        f"({row['external_pct']:.1f}%); "
        f"OR = {row['or']:.2f}, 95% CI [{row['or_lcl']:.2f}, {row['or_ucl']:.2f}], "
        f"p = {row['fisher_p']:.4g}"
    )

    if pd.notna(row["nnh"]):
        txt += (
            f"; NNH = {row['nnh']:.2f}, "
            f"95% CI [{row['nnh_lcl']:.2f}, {row['nnh_ucl']:.2f}]"
        )

    print(txt + ".")
```
